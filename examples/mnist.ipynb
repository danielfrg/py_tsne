```python
import gzip
import pickle
import numpy as np
from matplotlib import pyplot as plt

from tsne import bh_sne
```

```python
f = gzip.open("mnist.pkl.gz", "rb")
train, val, test = pickle.load(f, encoding="latin1")
f.close()
```

```python
X = np.asarray(np.vstack((train[0], val[0], test[0])), dtype=np.float64)
```

```python
y = np.hstack((train[1], val[1], test[1]))
```

**Note**: This one takes a while! Look at the terminal output where the Jupyter Notebook is running to see the progress from `verbose=True`.

```python
X_2d = bh_sne(X, verbose=True)
```

```python
plt.rcParams["figure.figsize"] = 20, 20
```

```python
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y)
```
